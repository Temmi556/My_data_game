```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sp
f = lambda x: x**2
df = lambda x: 2*x
x0 = 2

x = np.linspace(-2, 4, 200)
sns.lineplot(x=x, y=f(x))
plt.scatter(x0, f(x0))
plt.plot(x, f(x0) + df(x0)*(x - x0), '--')
plt.plot(x, (x*0), '-') # ось абсцисс
plt.show()
```
