```python
def np_t(first_array, second_array):
    b = np.ones((len(first_array),2))
    for i in range(len(first_array)):
        b[i,0] = first_array[i]
        b[i,1] = second_array[i]
    return b
np_t(np.array([1,2,3]), np.array([4,5,6]))
```
